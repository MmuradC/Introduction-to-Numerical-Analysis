# **Numerical Analysis: An Overview**

## What is a Number?
A number is a mathematical object used to count, measure, and label. The most basic examples are the **natural numbers** (e.g., 1, 2, 3, 4, ...). 

## What is an Equation?
In mathematics, an equation is a mathematical formula that expresses the equality of **two expressions**, by connecting them with the **equals sign =**.
Example (Polinomal Eq. (Parabola)):

$ Ax^2 + Bx + C - y = 0 $

## What is Numerical Analysis?
Numerical analysis is a branch of **mathematics** and **computer science** that focuses on developing **algorithms** and **computational methods** to solve mathematical problems numerically. Unlike symbolic or exact analytical solutions, numerical analysis deals with **approximations** to tackle problems that are too complex or impossible to solve by hand. Key areas include:
- **Error analysis**: Understanding and minimizing errors from approximations.
- **Convergence**: Ensuring solutions approach the correct result.
- **Stability**: Preventing small errors from growing uncontrollably.

## Why Analyze Numbers?
1. **Solving Real-World Problems**: Many practical problems (e.g., differential equations, optimization) lack exact solutions. Numerical methods approximate answers efficiently.
2. **Error Management**: All numerical computations introduce errors (e.g., rounding, truncation). Numerical analysis quantifies and mitigates these errors.
3. **Algorithm Efficiency**: Balancing speed and accuracy is critical in large-scale computations (e.g., simulations, big data).
4. **Computational Limitations**: Digital computers use finite precision (e.g., floating-point arithmetic). Numerical analysis adapts methods to these constraints.

### Applications of Numerical Analysis
- Engineering (structural simulations),
- Physics (climate modeling),
- Finance (option pricing),
- Machine learning (optimizing neural networks),
- Computer graphics (rendering algorithms).

# **Steps to Solve a Math Equation Using Numerical Analysis**

## 1. **Problem Definition**
   - **Understand the equation type**: Identify whether it is linear, nonlinear, differential, integral, or a system of equations.
   - **Define constraints**: Specify boundary conditions, initial values, or tolerance levels (e.g., acceptable error margins).

## 2. **Mathematical Formulation**
   - **Rewrite the problem** in a form suitable for numerical methods:
     - Root-finding: \( f(x) = 0 \)
     - Optimization: Minimize/Maximize \( f(x) \)
     - Linear systems: \( Ax = b \)
     - Differential equations: \( \frac{dy}{dt} = f(t, y) \)

## 3. **Choose a Numerical Method**
   - Select an algorithm based on the problem type and constraints:
     - **Linear equations**: Gaussian elimination, LU decomposition.
     - **Nonlinear equations**: Newton-Raphson, Bisection method.
     - **Differential equations**: Euler’s method, Runge-Kutta.
     - **Optimization**: Gradient descent, Newton’s method.
   - Consider factors like **convergence speed**, **stability**, and **computational cost**.

## 4. **Error Analysis**
   - **Identify error sources**:
     - **Truncation error**: Due to approximations (e.g., Taylor series truncation).
     - **Rounding error**: From finite-precision arithmetic in computers.
   - Estimate error bounds to ensure results are within acceptable tolerance.

## 5. **Implement the Algorithm**
   - **Code or compute iteratively**:
     - Use programming tools (Python, MATLAB) or manual calculations.
     - Example: For \( f(x) = x^2 - 2 = 0 \), apply the Newton-Raphson method:
       \[
       x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)} = x_n - \frac{x_n^2 - 2}{2x_n}
       \]

## 6. **Check Convergence**
   - Monitor if the solution approaches the true value:
     - Stop iterations when \( |x_{n+1} - x_n| < \text{tolerance} \).
     - If divergent, adjust the method or initial guess.

## 7. **Stability and Efficiency**
   - Ensure small errors do not amplify (e.g., avoid ill-conditioned matrices).
   - Optimize computational resources (time, memory).

## 8. **Validate and Verify**
   - **Validate**: Compare with analytical solutions (if available) or real-world data.
   - **Verify**: Test with simpler cases or known benchmarks.

## 9. **Document and Iterate**
   - Record the method, steps, and results.
   - Refine the approach if higher accuracy or efficiency is needed.

## Example: Solving \( e^x + x = 5 \)

1. **Formulate as root-finding**:  
   Define \( f(x) = e^x + x - 5 \).  
   Goal: Find \( x \) such that \( f(x) = 0 \).

2. **Choose method**: Newton-Raphson.  
   Iterative formula:  
   $ x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)} $

3. **Iterate**:  
   - **Initial guess**: \( x_0 = 1 \).  
   - **Update rule**:  
     $ x_{n+1} = x_n - \frac{e^{x_n} + x_n - 5}{e^{x_n} + 1} $

4. **Check convergence**:  
   Stop when \( |x_{n+1} - x_n| < 10^{-6} \).  
   Typical result: Converges to \( x \approx 1.30656 \).

# **Getting Started with Python for Numerical Analysis**

1. **NumPy** (`numpy`)  
   - **Purpose**: Efficient array operations, linear algebra, and numerical computations.  
   - **Key Features**:  
     - `ndarray` for fast multi-dimensional arrays.  
     - Broadcasting, vectorization, and mathematical functions (e.g., `np.sin`, `np.linalg.solve`).  
   - **Use Case**: Matrix operations, solving linear systems, and numerical simulations.

2. **SciPy** (`scipy`)  
   - **Purpose**: Advanced scientific computing (builds on NumPy).  
   - **Key Modules**:  
     - `scipy.integrate` (numerical integration),  
     - `scipy.optimize` (optimization),  
     - `scipy.interpolate` (interpolation).  
   - **Use Case**: Solving differential equations, curve fitting, and signal processing.

3. **Math** (`math`)  
   - **Purpose**: Basic mathematical functions (part of Python’s standard library).  
   - **Key Features**: Trigonometric, logarithmic, and constants like `math.pi`.  
   - **Use Case**: Simple calculations where NumPy is unnecessary.

4. **SymPy** (`sympy`)  
   - **Purpose**: **Symbolic mathematics** (exact algebraic manipulations).  
   - **Key Features**:  
     - Define symbolic variables (`symbols`),  
     - Solve equations symbolically (`solve`),  
     - Simplify expressions (`simplify`).  
   - **Use Case**: Deriving analytical formulas before numerical implementation.

5. **Matplotlib** (`matplotlib.pyplot`)  
   - **Purpose**: Data visualization.  
   - **Key Features**:  
     - Plot functions, histograms, and 3D graphs.  
     - Customize labels, legends, and styles.  
   - **Use Case**: Visualizing results of numerical analysis.

6. **Pretty Print** (`pprint`)  
   - **Purpose**: Format complex data structures (e.g., matrices, dictionaries) for readability.  
   - **Use Case**: Debugging and displaying symbolic expressions cleanly.
